```python
#Import Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the dataset
df=pd.read_csv('hotel_booking.csv')
```

```python
#This is used to generate descriptive statistics of a DataFrame
df.describe()
```

```python
#Displaying columns: Columns is not a method it is attribute of pandas dataframe
df.columns
```

```python
#Converting the reservation_status_date column to datetime 
df['reservation_status_date']=pd.to_datetime(df['reservation_status_date'])
```

```python
#This provides concise summary of a DataFrame
df.info()
```

```python
# Provides summary statistics for the string (object type) columns in a DataFrame.
df.describe(include='object')
```

```python
#For each string (object type) column in df, it prints the column name and its unique values.
for col in df.describe(include='object').columns:
    print(col)
    print(df[col].unique())
```

```python
#Getting null values in number
df.isnull().sum()
```

```python
#The code removes the 'company' and 'agent' columns from df 
#Drops rows with any missing values, updating the original DataFrame in both steps.
df.drop(['company','agent'],axis=1, inplace= True)
df.dropna(inplace=True)
```

```python
#Getting null values in number
df.isnull().sum()
```

```python
#This is used to generate descriptive statistics of a DataFrame
df.describe()
```

```python
#It filters the DataFrame df to retain only rows where the 'adr' column has values less than 5000.
df=df[df['adr']<5000]
```

```python
#This is used to generate descriptive statistics of a DataFrame
df.describe()
```

```python
#Data Analysis and Visualizations

#It computes the percentage distribution of unique values in the 'is_canceled' column of df and Assigns it to cancelled_perc
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

#It displays a bar chart with title 'Reservation_status_count' showing the count of 'Not canceled' vs 'Canceled' reservations from the 'is_canceled' column of df, with specific dimensions and styling.
plt.figure(figsize=(5,4))
plt.title('Reservation_status_count')
plt.bar(['Not canceled', 'Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
#It creates a bar plot with 'hotel' on the x-axis and counts on the y-axis, differentiating between 'Canceled' and 'Not Canceled' reservations, 
#styled with a blue palette, specific dimensions, and detailed labeling.
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue='is_canceled', data=df, palette='Blues')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(loc='upper right', bbox_to_anchor=(1,1))
plt.title('Reservation status in different hotels',size=20)
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

```python
#Selects and returns the percentage distribution of cancellations for entries with 'Resort Hotel' 
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#Selects and returns the percentage distribution of cancellations for entries with 'City Hotel' 
city_hotel=df[df['hotel']=='City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
#For both resort_hotel and city_hotel, it groups by 'reservation_status_date' and calculates the average 'adr' for each date
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
#Creates a plot of average daily rate (adr) over time for both City and Resort Hotels with specified labels and formatting
plt.figure(figsize=(20,8))
plt.title('Average Daily rate in City and Resort Hotel', fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label='Resort Hotel')
plt.plot(city_hotel.index,city_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
#Creates a plot showing the monthly count of reservations, distinguishing between canceled and not canceled, using 'reservation_status_date' in the DataFrame 'df'.
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(16,8))
ax1=sns.countplot(x='month', hue='is_canceled',data=df,palette='bright')
legend_labels,_=ax1.get_legend_handles_labels()
ax1.legend(loc='upper right', bbox_to_anchor=(1,1))
plt.title('Reservation status per month',size=20)
plt.xlabel('Month')
plt.ylabel('Number of reservations')
plt.legend(['Not Canceled','Canceled'])
plt.show()
```

```python
#Displays a bar chart of the summed monthly average daily rate ('adr') for canceled reservations, titled "ADR per month",using Seaborn.
plt.figure(figsize=(15,8))
plt.title('ADR per month',size=20)
data_to_plot = df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
sns.barplot(x='month', y='adr', data=data_to_plot)
plt.legend(fontsize=20)
plt.show()
```

```python
#Plots a pie chart of the top 10 countries with the highest number of canceled reservations from the DataFrame 'df'.
cancelled_data=df[df['is_canceled']==1]
top_10_country=cancelled_data['country'].value_counts()[:10]
plt.figure(figsize=(8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct= '%.2f' , labels=top_10_country.index)
plt.show()
```

```python
#Returns the count of each unique value in the 'market_segment' column of the 'df' DataFrame.
df['market_segment'].value_counts()
```

```python
#Returns the proportion (percentage) of each unique value in the 'market_segment' column of the 'df' DataFrame
df['market_segment'].value_counts(normalize=True)
```

```python
#Returns the proportion (percentage) of each unique value in the 'market_segment' column of the 'cancelled_data' DataFrame
cancelled_data['market_segment'].value_counts(normalize=True)
```

```python
#Computes the average daily rate ('adr') for each 'reservation_status_date' in the 'cancelled_data' dataset, then resets the index and sorts the result by date.
cancelled_df_adr=cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace=True)
cancelled_df_adr.sort_values('reservation_status_date',inplace=True )

#Filters the 'df' dataset for non-canceled reservations, then computes the average daily rate ('adr') by 'reservation_status_date', followed by resetting the index and sorting by date.
not_cancelled_data=df[df['is_canceled']==0]
not_cancelled_df_adr=not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace=True)
not_cancelled_df_adr.sort_values('reservation_status_date',inplace=True )

#Plots the average daily rate ('adr') over time for both canceled and non-canceled reservations, with respective labels, on a graph.
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend()

```

```python
#Filters both cancelled_df_adr and not_cancelled_df_adr datasets to only include entries with 'reservation_status_date' between 2017 and September 2017.
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016')&(cancelled_df_adr['reservation_status_date']<'2017-09') ]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016')&(not_cancelled_df_adr['reservation_status_date']<'2017-09') ]
```

```python
#Displays a graph of the average daily rate ('adr') over time for both canceled and non-canceled reservations with designated labels, using a title and legend.
plt.figure(figsize=(20,6))
plt.title('Average Daily Rate', fontsize=30)
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.legend(fontsize=20)
```
